```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead
```

```python
tokenizer=AutoTokenizer.from_pretrained('t5-base')
model=AutoModelWithLMHead.from_pretrained('t5-base',return_dict=True)
```

```python
Amazone_file = 'AmazoneT&C.txt'

# Define chunk size
chunk_size = 512

# Read the document
with open(Amazone_file, 'r') as file:
    essay = file.read()
```

```python
chunk_size=512
chunks=[essay[i:i+chunk_size] for i in range(0,len(essay),chunk_size)]
```

```python
chunk_summaries=[]
```

```python
for chunk in chunks:
    inputs = tokenizer.encode("summarize: " + chunk, return_tensors='pt', max_length=512, truncation=True)
    outputs = model.generate(inputs, max_length=300, min_length=100, length_penalty=5., num_beams=2)
    summary = tokenizer.decode(outputs[0])
    chunk_summaries.append(summary)
```

```python
unique_chunk_summaries = list(set(chunk_summaries))
```

```python
final_summary = " ".join(unique_chunk_summaries)
```

```python
final_summary
```

```python
inputs_final = tokenizer.encode("summarize: " + final_summary, return_tensors='pt', max_length=512, truncation=True)
outputs_final = model.generate(inputs_final, max_length=600, min_length=80, length_penalty=5., num_beams=2)
Legal_summary = tokenizer.decode(outputs_final[0])
```
